```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5,6])
y=np.array([12,22,34,48,56,69])
plt.scatter(x,y)
```

```python
def grad_desc(x,y):
    lr=0.08
    n=len(x)
    t0=t1=0
    T0=[]
    T1=[]
    itera=10000
    for i in range(itera):
        y_pred=t0+t1*x
        cost_func=(1/n)*sum(val**2 for val in (y-y_pred))
        td0=(2/n)*(sum((y_pred-y)))
        td1=(2/n)*(sum(x*(y_pred-y)))
        t0=t0-(lr/2)*td0
        t1=t1-(lr/2)*td1
        T0.append(t0)
        T1.append(t1)
#         J.append(cost_func)
        print("t0:{} t1:{} ct:{}".format(t0,t1,cost_func))
    X, Y = np.meshgrid(T0, T1)
    Z=X+x*Y
    plt.contourf(X, Y, Z)
grad_desc(x,y)
```
